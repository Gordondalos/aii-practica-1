```
import pandas as pd
import face_recognition
```

```
def detectFacesAndAddToDf(url):
  '''
  Функция распознавания лиц: выделения лиц на фото в баундинг-боксы
  вход: путь к изображению
  выход: список кортежей координат боксов распознанных лиц, список кортежей координат центров боксов распознанных лиц
  '''

  faces = face_recognition.load_image_file(url)
  facesLocations = face_recognition.face_locations(faces)
 
  facesLocationsCenters = []
  for (y_topright, x_topright, y_bottomleft, x_bottomleft) in facesLocations:
    center_y = (y_topright + y_bottomleft) / 2
    center_x = (x_topright + x_bottomleft) / 2
    centers = (center_y, center_x)
    facesLocationsCenters.append(centers)

  return facesLocations, facesLocationsCenters 
```

```
# Пример использования
data_yr = []
data_xr = []
for i in range(30):
  locs, centers = detectFacesAndAddToDf('/content/dataset-v1.1/val/images/' + str(5500+i) + '.jpg')
  if len(centers) == 0:
    data_yr.append(0)
    data_xr.append(0)    
  else:
    data_yr.append(centers[0][0])
    data_xr.append(centers[0][1])
  
df = pd.DataFrame(data_yr, data_xr)
```
